```python
#formula for closing balance
# ob = opening balance, add_mc = add material cost, sub_mc = substract material cost after term gets completed.
# payment = paytment done in the current month
def closing_balance(ob,add_mc,payment,sub_mc=0):
    cb = ob + add_mc - payment - sub_mc 
    return cb
```

```python
# date  = date when term has started.
# term = no of days in the term
# expected _date is the end date of the term
def expected_date(date,term):
    exp_date = pd.to_datetime(date) + timedelta(days=term)
    return exp_date
```

```python
l1 =[]

a = list(cash['Date'])
b = list(cash['payment_term']) 
for i,j in zip(a,b):
    l1.append(expected_date(i,j))
    
cash['expected_date'] = l1
```

```python
cash['month_year'] = pd.to_datetime(cash['Date']).dt.to_period('M')
cash['exp_month_year'] = pd.to_datetime(cash['expected_date']).dt.to_period('M')
```
