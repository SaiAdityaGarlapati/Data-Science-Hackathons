```python
import re 
  
def extractMax(input):  
     numbers = re.findall('\d+',input)  
     numbers = map(int,numbers) 
     return int(max(numbers))
```

```python
k=[]
for i in doctor['Experience']:
    k.append(extractMax(i))
```

```python
doctor['Exp_years'] = k
```

```python
k=[]
c=0
for i in doctor['Rating']:
    k.append(extractMax(str(i)))
    if extractMax(str(i)) ==0:
        c=c+1
        
doctor['Rating'] = k
```
